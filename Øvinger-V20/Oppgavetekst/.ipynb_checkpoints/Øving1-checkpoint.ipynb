Lager null- og ener-arrays som i oppgaveteksten:

Hvis numpy ikke hadde de gitte funksjonene ville det blitt reist feilmelding. Kan sjekke om funksjonene finnes uten å kræsje Python ved try-except. Bruker også variablen *pi* for å skrive mindre

```python
pi = np.pi
try:
    print(np.sin(pi), np.cos(0), np.tan(pi/4))
except:
    print("Noe har gått galt, sannsynligvis er numpy importert feil!") 
```

Svarene ser rare ut på grunn av avrundingsfeil, uttrykkene skal i realiteten bli 0, 1 og 1

Til slutt lager vi lineært space mellom 0 og $2 \pi$ på 100 punkter, og plotter trig-funksjonene:

```python
x = np.linspace(0, 2*pi, 100)
xAlt = np.logspace(0, 2*pi, 100)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.grid()
plt.show()
```

**Oppgave 2**

```python
#Importerer ikke numpy på nytt da den allerde er importert
def f(x):
    return 0.5+2/5*x-np.exp(-16*x**2)
x = np.linspace(-2,2,100)
plt.plot(x, f(x))
plt.grid()
plt.show()
```

Ser ut som $f(x)$ har $3$ nullpunkt

```python
def inthalv(f,a,b,tol):
    iterations = 0
    while (b-a)/2 > tol:
        m = 1/2*(a+b)
        iterations += 1
        if f(m)==0:
            return m, iterations
        elif f(a)*f(m) < 0:
            b=m
        elif f(a)*f(m) > 0:
            a=m
    return m, iterations
root, iterations = inthalv(f, -1.5, 1.5, 10**(-7))
if root > 0.186134:
    størrelse = "større"
else:
    størrelse = "mindre"
print("Roten er", np.abs(root-0.186134), størrelse, "enn 0.186134")
print("Funksjonen brukte", iterations, "intervallhalveringer")
```
