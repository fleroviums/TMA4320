Showing that (1)

$$
\phi(x,t)=\frac{1}{\sqrt{4\pi Dt}} \int_{-\infty}^\infty \mathrm{d}y \, h(y)e^{-\frac{(x-y)^2}{4Dt}}
$$

is a solution of the diffusion equation (2)

$$
\frac{\partial \phi(x,t)}{\partial t} =  D \frac{\partial^2 \phi(x,t)}{\partial x^2}
$$

Differentiating (1) with respect to time $t$

$$
\frac{ \partial \phi(x,t)}{\partial t} = \frac{ \partial}{\partial t} \left[\frac{1}{\sqrt{4\pi Dt}} \int_{\text{-}\infty}^\infty \mathrm{d}y \, h(y)e^{\,\text{-}\frac{(x-y)^2}{4Dt}} \right]
$$

where  $\left(4\pi D\right)^{\,\text{-}1/2}$ and $h(y)$ is independent of time. We simplify to

$$
\frac{ \partial \phi(x,t)}{\partial t} = \frac{ \partial}{\partial t} \left[\frac{1}{\sqrt{t}} e^{\,\text{-}\frac{(x-y)^2}{4Dt}} \right]  \int_{\text{-}\infty}^\infty \mathrm{d}y \, \frac{1}{\sqrt{4\pi D}}\, h(y)
$$

and can hereby use product rule of derivation. We define

$$
u \equiv \frac{1}{\sqrt{t}},
$$

$$
v \equiv e^{\text{-} \frac{(x-y)^2}{4Dt}},
$$

and

$$
A \equiv \int_{\text{-}\infty}^\infty \mathrm{d}y \, \frac{1}{\sqrt{4\pi D}} \, h(y),
$$

Where $A$ is independent of time. Thus, the derivation simplifies to

$$
\frac{ \partial \phi(x,t)}{\partial t} = \, \left[ \frac{\partial u}{\partial t} v \, + \, \frac{\partial v}{\partial t} u \right] \cdot A
$$

with the following solutions

$$
\frac{\partial u}{\partial t} = \frac{\text{-}1}{2 \, t^{\,3/2}},
$$

$$
\frac{\partial v}{\partial t} = \frac{(x-y)^2}{4Dt^2} e^{\,\text{-}\frac{(x-y)^2}{4Dt}}
$$

Thus, we have obtained the following expression for the substance distribution differentiated with respect to time

$$
\frac{ \partial \phi(x,t)}{\partial t} = \,   \left[ \frac{\text{-}1}{2 \, t^{\,3/2}} e^{\text{-} \frac{(x-y)^2}{4Dt}} \, + \, \frac{(x-y)^2}{4Dt^2} e^{\,\text{-}\frac{(x-y)^2}{4Dt}} \frac{1}{\sqrt{t}} \right] \cdot  \,A
$$

which can be simplified to

$$
\frac{ \partial \phi(x,t)}{\partial t} = \, \left(\frac{(x-y)^2}{4Dt^2}  \, + \, \frac{\text{-}1}{2 t}  \right) e^{\,\text{-}\frac{(x-y)^2}{4Dt}} \, \int_{\text{-}\infty}^\infty \mathrm{d}y\,\frac{1}{\sqrt{4\pi Dt}} \, h(y)
$$



The term of the second derivative of the distribution with respect to position similarly becomes

$$
D \frac{\partial^2 \phi(x,t)}{\partial x^2} = \frac{1}{4Dt^2} \left( \left(x-y\right)^2 - 2Dt \right) e^{\text{-}\frac{(x-y)^2}{4Dt}} \int_{\text{-}\infty}^\infty \mathrm{d}y\,\frac{1}{\sqrt{4\pi Dt}} \, h(y)
$$

and the equation is a solution.

We have the probability of finding a particle at position $x$ in a potential landscape $V(x)$ given by

$$
P(x) = Z e^{-\beta V(x)}, 
$$

and hereby find the probability of moving one step to the right $p^+$

$$
p^+ = \frac{P(x_0 + 1)}{P(x_0) + P(x_0 + 1) + P(x_0 -1)},
$$

$$
\quad \quad \, \, \, \, \quad = \frac{Z e^{-\beta V(x_0 + 1)}}{Z e^{-\beta V(x_0)} + Z e^{-\beta V(x_0 + 1)} + Z e^{-\beta V(x_0 - 1)}},
$$

which can be simplified to

$$
p^+ = \frac{1}{1 + e^{-\beta [V(x_0) - V(x_0+1)]} + e^{-\beta [V(x_0 - 1)-V(x_0+1)]}}.

Similarly, we have the probailities of staying put, $p^0$, and moving one step to the left, $p^-$ given by

$$

$$

$$
p^- = \frac{1}{1 + e^{-\beta [V(x_0) - V(x_0-1)]} + e^{-\beta [V(x_0 + 1)-V(x_0-1)]}},
$$

```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.stats import norm


particles = 10000
#beta = 1/(1.38064852*10**(-23)*298)
kb = [0.01, 1, 100] #k * beta
x = np.linspace(-200,200,400)
T = 200 #tidsintervall

def p_minus(x, kb, V):
    return 1/(1 + math.exp(-kb*(V(x+1)-V(x-1)))+math.exp(-kb*(V(x)-V(x-1))))

def p_zero(x, kb, V): #Unused, may be removed
        return 1/(1 + math.exp(-kb*(V(x-1)-V(x)))+math.exp(-kb*(V(x+1)-V(x))))

def p_plus(x, kb, V):
    return 1/(1 + math.exp(-kb*(V(x-1)-V(x+1)))+math.exp(-kb*(V(x)-V(x+1))))

def rand_walk(kb, T, V):
    x = 0 #x0
    left = 0
    right = 0
    for i in range(0, T):
        rNum = random.random()
        if (rNum) <= p_minus(x, kb, V):
            x-=1 #Move 1 step to the left
            #left += 1
            #print("venstre")
        elif (rNum) > (1-p_plus(x, kb, V)):
            x+=1 #Move 1 step to the right
            #print("høyre")
            #right += 1
        else:
            x+=0 #Does nothing but for readability
    return x

def normalPlot(points):
    mu, std = norm.fit(points)
    p = norm.pdf(x,mu,std)
    plt.plot(x,p,'k')
    plt.hist(points, bins=10, density=True, edgecolor='blue', color='cornflowerblue')
    #(unique, counts) = np.unique(points, return_counts=True)
    #counts = counts/particles #Scales points relative to normal distribution
    #plt.plot(unique, counts, 'o', linewidth=None)
    plt.show()
```

```python
def V_const(x):
    return 1 #Defining V(x)/k in reality

xvals = np.zeros(10000) #Empty np-set


for i in range(int(particles)):
        xvals[i] = rand_walk(kb[0], T, V_const)#.x
plt.title("$V=k,\, kb=0.01$")
plt.ylim(top=0.04)  # adjust the top leaving bottom unchanged
normalPlot(xvals)

#kb=1
for i in range(int(particles)):
    xvals[i] = rand_walk(kb[1], T, V_const)#.x det funker vel ikke.. + ugunstig

plt.title("$V=k,\, kb=1$")
normalPlot(xvals)


#kb=100
for i in range(int(particles)):
    xvals[i] = rand_walk(kb[2], T, V_const)
plt.title("$V=k,\, kb=100$")
normalPlot(xvals)
```

**Comments regarding $V=k$:**
The above graphs illustrate the particle distribution and its normal fit curve when the particles are in one dimension and subject to a constant potential $V=k$. These illustrations are based on $T=200$ time steps. We observe that the different $\beta k$-values do not affect the normal distribution curve whatsoever.

In general, $\beta k$ is the ratio of $k$ to $k_B T$, which in this case is a ratio that tells us something about how the distribution of the particles are influenced by the potential they are in and the thermal energy of the particles.
It can also be interpreted from the equations (3), (4), and (5) that a constant potential will lead the power $[V(x+1)-V(x)]=0$, so that the potential that the particles are subject to, will *not* influence their distribution.

Hence, one can conclude that, in a constant potential field, the thermal energy of the particles dominates the probability distribution of the particle motion. However, since the distributions are marginally influenced by the change in $\beta k$, the thermal energy of the particles is seen to be insignificant on its impact to the probability distribution of the particle motion. Concludingly, $\beta k$ cannot be used as a ratio to determine whether the potential or thermal energy dominates the random walk.

Since the potential increases proportionally with $x$, the particle motion will tend towards $-x$ which is defined to be leftwards. Moreover, increasing $\beta k$ is equivalent to reducing the thermal energy of the particles. Hence, the thermal energy has negligible contribution to the particle distribution, and it is said that the potential that the particles are subject to, dominates their distribution.

```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import time
from numba import jit

b= 1000 
k=1
a = 0.8
N_x = 90
h = 1
a = 0.8

@jit
def p_pluss5(x,pot):
    return 1/(1 + np.exp(-b*(pot[x-1]-pot[x+1]))+np.exp(-b*(pot[x]-pot[x+1])))
@jit
def p_minus5(x,pot):
    #x = int(x)
    return 1/(1 + np.exp(-b*(pot[x+1]-pot[x-1]))+np.exp(-b*(pot[x]-pot[x-1])))

# Defining the two potential functions in terms of lists
Na = 2
x1 = 2*N_x
x = np.linspace(0,180,180)
V1 = signal.sawtooth(x/x1*2*np.pi*Na, width=a)
V2 = np.ones(180) #Arbitrary constant function
####


@jit
def rtime(pot,x):
    h = 1
    N_x = 90
    rand_walk = random.random()
    if rand_walk <= p_minus5(x, pot):
        return -1, -1 #Step to left
    elif rand_walk > (1-p_pluss5(x,pot)):
        return 1, 1 #Step to right
    else:
        return 0, 0

@jit
def cycle():
    t_inter = 1000 #should be 1000=2*500
    startTime = time.time()
    cycles = 10
    j_ti = []
    x = 0 #start position 0
    J=0
    N_p=12*90 #ant partikler 
    position = np.floor(np.linspace(0,180,num=1080,endpoint=False)) #posisjonen til alle partiklene
    for i in range(cycles):
        print("Cycle:", i)
        if i!=0:
            print("Last cycle runtime:", time.time()-cycleTime)
        cycleTime = time.time()
        J =0
        for h in range(t_inter):
            if h>= 500: #500, when to change potential
                potential = V1
            else:
                potential = V2
            for i in range(N_p):
                if position[i]>=179:
                    position[i] = 0
                elif position[i]<0:
                    position[i] = 178
                pos, current = rtime(potential, int(position[i])) #int
                position[i]+=pos
                J+=current
        one_cycle = J/N_p
        one_cycle = one_cycle/(t_inter)
        j_ti.append(one_cycle) #adding normalized particle current for each cycle
    print("Total runtime:", time.time()-startTime)
    return j_ti # J_avg

    
print("10 cycle for a=0.8:")
#print(cycle())
```
